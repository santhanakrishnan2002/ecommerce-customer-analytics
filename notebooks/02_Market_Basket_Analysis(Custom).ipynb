```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("b2c.csv")
df.head()
```

```python
Country_name =input("Enter The Country Name Here")
```

```python
df1 = str(Country_name)
```

```python
df1 = df[df["Country"]==df1]
```

```python
df1.head()
```

```python
df1.shape
```

```python
basket = df1.groupby(["Invoice", "Description"])["Quantity"].sum().unstack()
```

```python
basket = basket.fillna(0)
```

```python
def encode_units(x):
    if x >= 1:
        return True
    if x <= 0:
        return False
```

```python
basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.head()
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.02, use_colnames = True)
```

```python
print(f"Most Frequent itemsets:(support>=0)")
```

```python
frequent_itemsets.sort_values(by="support", ascending=False).head()
```

```python
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1.0)
```

```python
rules.sort_values(by = ["lift","support"], ascending = False, inplace = True)
```

```python
print(f"Top n/ most frequent itemsets")
```
